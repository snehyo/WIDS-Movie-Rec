```python
import pandas as pd
```

```python
df1=pd.read_csv("movies_metadata.csv")
df2=pd.read_csv("links.csv")
df3=pd.read_csv("keywords.csv")
df4=pd.read_csv("credits.csv")
```

```python
df1["id"] = pd.to_numeric(df1["id"], errors="coerce")
df1 = df1.dropna(subset=["id"])
df1["id"] = df1["id"].astype(int)
```

```python
df13=df1.merge(df3,on='id')
```

```python
df134=df13.merge(df4,on='id')
```

```python
df_final = df134[df134['id'].isin(df2['tmdbId'])]
```
