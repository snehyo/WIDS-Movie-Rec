```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df=pd.read_csv(r"C:\Users\snehm\OneDrive\Desktop\WIDS\master_dataset_final.csv")
```

```python
cv = CountVectorizer(
    analyzer='word',
    ngram_range=(1, 2), 
    min_df=2,             
    stop_words='english'
)
```

```python
count_matrix = cv.fit_transform(df['soup'])
```

```python
count_matrix.shape
```

```python
cosine_sim=cosine_similarity(count_matrix)
cosine_sim_df=pd.DataFrame(cosine_sim)
```

```python
cosine_sim_df.to_csv(r"C:\Users\snehm\OneDrive\Desktop\WIDS\cosine_similarity_matrix.csv")
```

```python
def recommend_movies(movie_title, top_n=5):
    if movie_title not in df['title'].values:
        return "Movie not found in dataset"

    idx = df[df['title'] == movie_title].index[0]

    similarity_scores = list(enumerate(cosine_sim[idx]))

    similarity_scores = sorted(
        similarity_scores,
        key=lambda x: x[1],
        reverse=True
    )

    similarity_scores = similarity_scores[1:top_n+1]

    recommendations = []

    for i, score in similarity_scores:
        recommendations.append({
            "title": df.iloc[i]['title'],
            "director": df.iloc[i]['director'],
            "release_date": df.iloc[i]['release_date']
        })

    return pd.DataFrame(recommendations)
```
